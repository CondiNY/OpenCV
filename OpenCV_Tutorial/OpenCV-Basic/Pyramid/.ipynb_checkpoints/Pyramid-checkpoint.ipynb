# 图像金字塔

## 1 API说明

- 下采样:pyrDown(src)
- 上采样:pyrUp(dst)

## 2 代码实现

### 2.1 上采样

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread('InputImage/goat.png')
plt.imshow(img)
```

```python
img.shape
```

```python
# 下采样
dst = cv2.pyrDown(img)
plt.imshow(dst)
```

```python
dst.shape
```

```python
dst = cv2.pyrDown(dst)
plt.imshow(dst)
```

```python
dst.shape
```

- 总结：通过下采样后每次下才有都会减小１/2倍,图像也变得模糊了

### 2.2 上采样

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread('InputImage/goat.png')
plt.imshow(img)
```

```python
img.shape
```

```python
# 上采样
src1 = cv2.pyrUp(img)
plt.imshow(src1)
```

```python
src1.shape
```

```python
# 上采样
src2 = cv2.pyrUp(src1)
plt.imshow(src2)
```

```python
src2.shape
```

- 通过以上的维度可以看出，每次上采样图像的维度都增加两倍．

## ３总结
- 注意：但须要注意的是。PryUp和PryDown不是互逆的。即PryUp不是降採样的逆操作。这样的情况下，图像首先在每一个维度上扩大为原来的两倍，新增的行（偶数行）以0填充。然后给指定的滤波器进行卷积（实际上是一个在每一个维度都扩大为原来两倍的过滤器）去预计“丢失”像素的近似值。
